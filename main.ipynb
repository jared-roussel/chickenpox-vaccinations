```python
#Read CSV file for 1st data source Vaccine Coverage and Exemptions among Kindergarteners
coverage_exmpt_file = "Data\Vaccination_Coverage_and_Exemptions_among_Kindergartners.csv"
coverage_exmpt_file_df = pd.read_csv(coverage_exmpt_file, encoding="ISO-8859-1")
coverage_exmpt_file_df.head(100)
```

```python
#Filter only for chickenpox vaccine (varicella)
chickenpox_df = coverage_exmpt_file_df.loc[coverage_exmpt_file_df['Vaccine/Exemption'] == 'Varicella, unknown disease history']
chickenpox_df.head(100)
```

```python
#Check Survey Types
#chickepox_df_filtered_test = chickenpox_df['Survey Type'].value_counts()
#chickepox_df_filtered_test 
```

```python
#Filter out Random Sample
chickepox_df_filtered_RS_df = chickenpox_df.loc[chickenpox_df['Survey Type'] != 'Random Sample']
chickepox_df_filtered_RS_df.head()
```

```python
#Filter out Not Conducted Survey Types
chickepox_df_filtered_NC_df = chickepox_df_filtered_RS_df.loc[chickepox_df_filtered_RS_df['Survey Type'] != 'Not Conducted']
chickepox_df_filtered_NC_df.head(100)
```

```python
chickepox_df_filtered_NaN_df = chickepox_df_filtered_NC_df.loc[chickepox_df_filtered_NC_df['Survey Type'] != 'NaN']
chickepox_df_filtered_NaN_df
```

```python
#Filter table for years 2016 - 2019
chickepox_df_filtered_NaN_df['Year'] = pd.to_numeric(chickepox_df_filtered_NaN_df['School Year'].str[:4])
#chickepox_df_filtered_NaN_df = chickepox_df_filtered_NaN_df.drop(columns=['School Year - Year'])
chickepox_df_filtered_NaN_df.reset_index()
chickepox_df_filtered_Years = chickepox_df_filtered_NaN_df.loc[(chickepox_df_filtered_NaN_df['Year']) >= 2016]
chickepox_df_filtered_Years
```

```python
#Drop unnecessary columns
chickepox_df_filtered_Years.reset_index(inplace =True)
#chickepox_df_filtered_Years.drop(columns='Index')
chickepox_df_filtered_Years.drop(columns='Footnotes')
chickepox_df_filtered_Years
```
